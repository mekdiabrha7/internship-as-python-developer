```python
#task 1
!pip install plyer
import notification
import time
from datetime import datetime

#  Define your reminders here
reminders = [
    {"task": "Math Club Meeting", "time": "2025-10-26 21:00"},
    {"task": "Python Project Deadline", "time": "2025-10-27 10:00"},
    {"task": "Morning Study Session", "time": "2025-10-27 06:00"}
]

print("ğŸ”” Reminder system started...")

while True:
    now = datetime.now().strftime("%Y-%m-%d %H:%M")

    for reminder in reminders:
        if reminder["time"] == now:
            notification.notify(
                title="â° Reminder!",
                message=f"It's time for: {reminder['task']}",
                timeout=10  # seconds
            )
            print(f"Reminder: {reminder['task']} at {now}")
            # Prevent duplicate notifications
            reminders.remove(reminder)
            break

    time.sleep(30)  # check every 30 seconds
```
